```python
def gd(training_data,learning_rate,number_of_iteration,n):
    m=[0 for i in range(n)]
    c=0
    for i in range(number_of_iteration):
        m,c=step_gradient(training_data,learning_rate,m,c,n)
        cost(training_data,m,c)
    return m,c
```

```python
def step_gradient(training_data,learning_rate,m,c,n):
    m_slope=[0 for i in range(n)]
    c_slope=0
    M=len(training_data)
    for t in range(n):
        for i in range(M):
            x=training_data[i,0:-1]
            y=training_data[i,-1]
            val=0
            for j in range(len(x)):
                val+=m[j]*x[j]
            m_slope[t]+=(-2/M)*((y-val-c)*x[t])
            c_slope+=(-2/M)*(y-val-c)
    for i in range(n):
        m[i]=m[i]-learning_rate*m_slope[i]
    c=c-learning_rate*c_slope
    return m,c
```

```python
def cost(training_data,m,c):
    total=0
    M=len(training_data)
    for i in range(M):
        x=training_data[i,0:-1]
        y=training_data[i,-1]
        val=0
        for j in range(len(x)):
            val+=m[j]*x[j]
        total+=(1/M)*((y-val-c)**2)
    return total
```

```python
def run():
    learning_rate=0.0099
    number_of_iteration=500
    m,c=gd(training_data,learning_rate,number_of_iteration,13)
    return m,c
```

```python
y_pred=[]
def predict(testing_data,m,c,n):
    M=len(testing_data)
    for i in range(M):
        x=testing_data[i,:]
        val=0
        for j in range(n):
            val+=m[j]*x[j]
        ans=val+c
        y_pred.append(ans)
```

```python
m,c=run()
predict(testing_data,m,c,13)
for i in y_pred:
    print(i)
```
